In the real problems, a lot of datasets in the security are not labeled, or when it is, it is not reliable. Because of this situation, an important task of a Data Analyst is to identify features in the dataset and, using probability, try to infer labels in to the dataset

**Using the provided unlabaled dataset, create a python method that calculates how likely is that a connection from IP '14.139.187.130' and the US can make a transaction at 17-01-01 06:56:17**

First we import the dataset

```python
import pandas as pd


def import_dataset(filename):
    ds = pd.read_csv(filename + '.csv',
                     parse_dates=[3], infer_datetime_format=True)
    ds['datetime'] = ds['datetime'].sort_index()
    return ds

import_dataset("access_log_format-1")
```

```python
ds = import_dataset("access_log_format-1")
```

**Now we get the transactions that that were made at 17-01-01 06:56:17**

```python
def get_rows_by_time(df, ts):
    dst = df[df['datetime'] == ts]
    return dst

get_rows_by_time(ds, '2017-01-01 06:56:17') 
```

```python
def get_rows_by_ip(df, ip):
    dst = df[df['src_ip'] == ip]
    return dst

get_rows_by_ip(ds, '14.139.187.130') 
```

```python
def get_rows_by_country(df, country):
    dst = df[df['country'] == country]
    return dst

get_rows_by_country(ds, 'US') 
```

In the data above we can see that at the datetime 2017-01-01 06:56:17 there is no value for neither US or ip: 14.139.187.130, so we can be almost sure that the intersection is 0 and therefore the probability is 0, now we prove it.

**We add boolean columns with the data to show if the entry has the data we want**

```python
import numpy as np

def prepare_pivot_table(ds):
    ds['From_US'] = np.where(ds['country'] == "US", 1, 0)
    ds['From_IP'] = np.where(ds['src_ip'] == "14.139.187.130", 1, 0)
    ds['Correct_datetime'] = np.where(ds['datetime'] == "2017-01-01 06:56:17", 1, 0)
    ds['count'] = 1
    
    df = ds[['From_IP', 'From_US', 'Correct_datetime', 'count']]

    return df

df = prepare_pivot_table(ds)
df.head()
```

```python
pd.pivot_table(
    df,
    values = 'count',
    index = ['From_IP'],
    columns = ['From_US', 'Correct_datetime'],
    aggfunc = np.size,
    fill_value = 0
)
```

Proving that there is no case in which the IP 14.139.187.130 and the US happen at the same time and neither of those values appear during the datetime 2017-01-01 06:56:17 we can now that the value of the intersection (US âˆ© 14.139.187.130 âˆ© 2017-01-01 06:56:17) is zero we then we know that: 

**0/P(2017-01-01 06:56:17)is zero**
